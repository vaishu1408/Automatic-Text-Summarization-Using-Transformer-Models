```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Path to the saved model
model_path = r"C:\Users\Welcome\OneDrive\Desktop\Text Analytics\Final Project"

# Load the tokenizer and model
tokenizer = T5Tokenizer.from_pretrained(model_path)
model = T5ForConditionalGeneration.from_pretrained(model_path)

# If using GPU, move the model to the correct device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)

# Example inference
input_text = "Environmental conservation is the practice of protecting and preserving the natural world. It includes efforts to maintain biodiversity, reduce pollution, protect ecosystems, and ensure sustainable resources for future generations. As human activities have escalated over the past few centuries, the need for environmental conservation has become more urgent. Deforestation, overfishing, pollution, and climate change are some of the key issues threatening the health of our planet. One of the most significant contributors to environmental degradation is deforestation. Trees play a critical role in absorbing carbon dioxide and producing oxygen, but when forests are destroyed for agriculture or urban development, the carbon stored in trees is released into the atmosphere, contributing to global warming. Additionally, the destruction of forests leads to habitat loss for countless species of animals and plants. Pollution is another major environmental issue. Air pollution, caused by industrial activities, car emissions, and burning fossil fuels, contributes to respiratory diseases and global warming. Water pollution, often resulting from industrial runoff and improper waste disposal, harms aquatic life and makes water unsafe for human consumption. Similarly, land pollution from plastic waste and other materials is a growing concern, with millions of tons of waste entering the oceans each year, harming marine life and threatening biodiversity. Climate change, driven largely by human activities such as the burning of fossil fuels and deforestation, has become one of the most pressing global challenges. Rising global temperatures have caused melting ice caps, sea-level rise, and more frequent extreme weather events such as hurricanes, droughts, and floods. These changes threaten agriculture, water resources, and biodiversity, affecting millions of people worldwide. In order to mitigate these issues, individuals, governments, and organizations must work together to promote sustainable practices. Reducing carbon emissions, transitioning to renewable energy sources, adopting eco-friendly practices, and supporting policies that protect natural habitats are essential steps toward environmental conservation. Education and raising awareness about the importance of conservation can also play a critical role in motivating positive change. Ultimately, the future of our planet depends on our collective efforts to conserve and protect the environment. By taking action today, we can ensure that future generations inherit a healthier, more sustainable world."
input_ids = tokenizer(input_text, return_tensors="pt").input_ids.to(device)

# Generate summary
summary_ids = model.generate(input_ids, max_length=64, min_length=10, length_penalty=2.0, num_beams=4)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print(f"Generated Summary: {summary}")
```
